Created a function to "One Hot Encode" the target outputs.

```python
def ohc(arry):
    if arry == 1:
        return [1,0,0,0,0,0]
    if arry == 2:
        return [0,1,0,0,0,0]
    if arry == 3:
        return [0,0,1,0,0,0]
    if arry == 4:
        return [0,0,0,1,0,0]
    if arry == 5:
        return [0,0,0,0,1,0]
    if arry == 6:
        return [0,0,0,0,0,1]
    
```

```python
train_labels = np.zeros(10000)
train_target = np.zeros(10000)
train_labels = train_labels.tolist()
train_target = train_target.tolist()
for i,train_img_name in enumerate(os.listdir(TRAIN_DIR)):
    train_labels[i] = ohc(int(train_img_name.split('.')[0].split('_')[-1]))
    print(train_img_name.split('.')[0].split('_')[-1])
    train_img_arr = cv2.imread(os.path.join(TRAIN_DIR,train_img_name))   
    train_target[i] =  train_img_arr
```
